DATA CLEANING PROCESS

```python
#Dropping unnecessary columns
crime=crime.drop(['Unnamed: 0','DISTRICT'],axis=1)
crime.head()
```

```python
def get_case_consistency(row):
    row=row['STATE/UT'].strip()
    row=row.upper()
    row=row.title()
    return row
crime['STATE/UT']=crime.apply(get_case_consistency,axis=1)
crime['STATE/UT'].unique()
```

```python
North_india=['Jammu & Kashmir','Himachal Pradesh','Punjab','Haryana','Uttar Pradesh','Uttarakhand','Chandigarh']
East_india=['Assam','Arunachal Pradesh','Manipur','Mizoram','Nagaland','Sikkim',"Meghalaya",'Tripura']
South_india=['Andhra Pradesh','Karnataka','Kerala','Tamil Nadu','Telangana']
West_india=['Rajasthan','Gujarat','Goa','Maharashtra','Goa']
Central_india=['Madhya Pradesh','Chhattisgarh']
North_east_india=['Bihar','Jharkhand','Odisha','West Bengal']
Ut_india=['A & N Islands','Delhi','LAKSHADWEEP','PUDUCHERRY','A & N ISLANDS','D & N Haveli','Daman & Diu','Delhi Ut','Lakshadweep','Puducherry',
          'DAMAN & DIU','D&N Haveli','A&N Islands']

def get_zonal_names(row):
    if row['STATE/UT'].title().strip() in North_india:
        val= 'North Zone'
    elif row['STATE/UT'].title().strip() in East_india:
        val= 'East Zone'
    elif row['STATE/UT'].title().strip() in South_india:
        val= 'South Zone'
    elif row['STATE/UT'].title().strip() in West_india:
        val='West Zone'
    elif row['STATE/UT'].title().strip() in Central_india:
        val= 'Central Zone'
    elif row['STATE/UT'].title().strip() in North_east_india:
        val= 'North East Zone'
    elif row['STATE/UT'].title().strip() in Ut_india:
        val= 'Union Territory'
    else:
        val= 'Other'
    return val
crime['Zones']=crime.apply(get_zonal_names,axis=1)
crime['Zones'].unique()
```

```python
crime[(crime['Zones']=='Other')]['STATE/UT'].unique()
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women',
        'Cruelty by Husband or his Relatives','Importation of Girls']
```

CRIMES IN DIFFERENT ZONES

RAPE CASES

```python
rape=crime.groupby(by=['Year','STATE/UT','Zones'])['Rape'].sum().reset_index().sort_values(crimes[0],ascending=False)
rape
```

REGISTERED RAPE CASES YEAR (2001-2014)

```python
plt.figure(figsize=(20,20))
count=1
for zone in rape.Zones.unique():
    plt.subplot(rape.Zones.unique().shape[0],1,count)   
    sns.lineplot(x='Year',y=crimes[0],data=rape[rape['Zones']==zone])
    count+=1
    plt.subplots_adjust(hspace=0.5)
    plt.xlabel('Year')
    plt.ylabel('Rape cases')
    plt.title(zone)
    plt.show()
```

```python
#Zonal registered rape cases

fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x='Zones', y='Rape', data=rape, errwidth=0, ax=ax,palette='Set2')
ax.set_ylabel('Rape cases')
ax.set_title('Zone-wise rape cases registered', fontdict={'fontsize': 15})
plt.xticks(rotation=45)
plt.show()
```

```python
#State wise registered rape cases
rape_state = rape[(rape['Zones'] == 'Central Zone') | (rape['Zones'] == 'East Zone')]
rape_state = rape_state.groupby(by=['STATE/UT'])['Rape'].sum().reset_index().sort_values('Rape', ascending=False)
fig, ax = plt.subplots(figsize=(15,15))
sns.barplot(x='STATE/UT', y='Rape', data=rape_state, errwidth=0, ax=ax,palette='Set1')
ax.set_ylabel('Rape cases')
ax.set_title('State wise registered rape cases', fontdict={'fontsize': 15})

rape_state.head(5)
```

```python
df=pd.read_csv("crimes_against_women_2001-2014.csv")
df.columns
```

```python
y1 = df['Year'].max()
print(y1)
y2 = y1 - 1
y3 = y2 - 1
y4 = y3 - 1

def pattern(a, b, c):
    if (a == 1):
        if (b == 1):
            if (c == 1): return "Higher chances of an increase"
            else: return "Medium chances of a decrease"
        elif (b == 0):
            if (c == 1): return "Medium chances of an increase"
            else: return "Lower chances of an increase"
    elif (a == 0):
        if (b == 0):
            if (c == 0): return "Lower chances of an increase"
            else: return "Medium chances of an increase"
        elif (c == 0):
            if (c == 1): return "Medium chances of a decrease"
            else: return "Higher chances of an increase"

x = input("Enter the State/UT: ")
y = input("Enter the crime: ")

# Extract values for the specified State/UT and years
x1 = list(df[(df['STATE/UT'] == x) & (df['Year'] == y1)][y])
x2 = list(df[(df['STATE/UT'] == x) & (df['Year'] == y2)][y])
x3 = list(df[(df['STATE/UT'] == x) & (df['Year'] == y3)][y])
x4 = list(df[(df['STATE/UT'] == x) & (df['Year'] == y4)][y])

# Initialize s1, s2, s3
s1, s2, s3 = 0, 0, 0

# Check if lists are not empty before accessing elements
if (len(x1) > 0 and len(x2) > 0):
    if (x == 'Telangana' and y1 == 2022):
        s1 = 0
        s2 = 0
        s3 = 1 if (x1[0] - x2[0]) > 0 else 0
    else:
        if (len(x3) > 0 and len(x4) > 0):
            s1 = 1 if (x3[0] - x4[0]) > 0 else 0
        if (len(x2) > 0 and len(x3) > 0):
            s2 = 1 if (x2[0] - x3[0]) > 0 else 0
        if (len(x1) > 0 and len(x2) > 0):
            s3 = 1 if (x1[0] - x2[0]) > 0 else 0
else:
    print(f"No data available for {x} in the year {y1}.")
    exit()

# Get the prediction result
res = pattern(s1, s2, s3)
print(f"{x} has {res} in {y} in the year {str(y1 + 6)}, considering constant current policies.")
```
